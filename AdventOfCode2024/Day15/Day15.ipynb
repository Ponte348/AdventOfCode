```python
def findRobot(grid):
    for y in range(len(grid)):
        for x in range(len(grid[y])):
            if grid[y][x] == '@':
                return (x, y)
    print('Robot not found')
    return None
robot_pos = findRobot(grid)
print(robot_pos)

directions = {"^": (0, -1), "v": (0, 1), "<": (-1, 0), ">": (1, 0)}

for move in movements:
    next_pos = (robot_pos[0]+directions[move][0], robot_pos[1]+directions[move][1])
    
    if grid[next_pos[1]][next_pos[0]] == '#':
        continue
    
    if grid[next_pos[1]][next_pos[0]] == 'O':
        # we found a box, find the next empty space in the direction of the move and move everything 1 step closer
        distance = 0
        while grid[next_pos[1]+directions[move][1]*distance][next_pos[0]+directions[move][0]*distance] != '.' and grid[next_pos[1]+directions[move][1]*distance][next_pos[0]+directions[move][0]*distance] != '#':
            distance+=1
            
        
        
        print(distance)
    
```
