PART 1

```python
f = open('input.txt', 'r')
alturas=[]
lin=0
vis=0
visedge=0
visivel=0
for row in f.read().splitlines(): # Read inputs
    alturas.append([])            # Add a new row !
    for height in row:            # To create 2D matrix with inputs
        alturas[lin].append(int(height)) # Add a new column !?
    lin+=1

for i in range(len(alturas)):
    for j in range(len(alturas[i])): # value = alturas[i][j]
        visivel=0
        if i==0 or i==len(alturas)-1 or j==0 or j==len(alturas[i])-1:
            visedge+=1
            continue
        else: # if not edge
            for k in range(j-1, -1, -1): # left
                if alturas[i][j]<=alturas[i][k]:
                    visivel=0
                    break
                else:
                    visivel=1
            if visivel:
                vis+=1
                continue
            
            for k in range(j+1, len(alturas[i])): # right
                if alturas[i][j]<=alturas[i][k]:
                    visivel=0
                    break
                else:
                    visivel=1
            if visivel:
                vis+=1
                continue

            for k in range(i-1, -1, -1): # up
                if alturas[i][j]<=alturas[k][j]:
                    visivel=0
                    break
                else:
                    visivel=1
            if visivel:
                vis+=1
                continue
            
            for k in range(i+1, len(alturas)): # down
                if alturas[i][j]<=alturas[k][j]:
                    visivel=0
                    break
                else:
                    visivel=1
            if visivel:
                vis+=1
                continue

print(vis+visedge)
f.close()
```

```python
f = open('input.txt', 'r')
alturas=[]
lin=0
vis=0
visedge=0
visivel=0
maxscenic=0

for row in f.read().splitlines(): # Read inputs
    alturas.append([])            # Add a new row !
    for height in row:            # To create 2D matrix with inputs
        alturas[lin].append(int(height)) # Add a new column !?
    lin+=1

for i in range(len(alturas)):
    for j in range(len(alturas[i])): # value = alturas[i][j]
        visis=[0,0,0,0]
        scenic=0
        visivel=0
        if i==0 or i==len(alturas)-1 or j==0 or j==len(alturas[i])-1:
            visedge+=1
            continue
        else: # if not edge
            for k in range(j-1, -1, -1): # left
                visis[0]+=1
                if alturas[i][j]<=alturas[i][k]:
                    visivel=0
                    break
                else:
                    visivel=1
            if visivel:
                vis+=1
            
            for k in range(j+1, len(alturas[i])): # right
                visis[1]+=1
                if alturas[i][j]<=alturas[i][k]:
                    visivel=0
                    break
                else:
                    visivel=1
            if visivel:
                vis+=1

            for k in range(i-1, -1, -1): # up
                visis[2]+=1
                if alturas[i][j]<=alturas[k][j]:
                    visivel=0
                    break
                else:
                    visivel=1
            if visivel:
                vis+=1
            
            for k in range(i+1, len(alturas)): # down
                visis[3]+=1
                if alturas[i][j]<=alturas[k][j]:
                    visivel=0
                    break
                else:
                    visivel=1
            if visivel:
                vis+=1
            
            # multiply every num in visis
            scenic=1
            for num in visis:
                scenic*=num
            # if its bigger than maxscenic, maxscenic = scenic
            if scenic>maxscenic:
                maxscenic=scenic
    
print(maxscenic)
f.close()
```
