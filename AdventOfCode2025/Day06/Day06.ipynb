```python
from time import perf_counter_ns
```

## Part 1

```python
start_time = perf_counter_ns()

with open("input.txt") as f:
    data = [line.split() for line in f.read().strip().splitlines()]

calculations = 0

for i in range(len(data[0])):  # for each column
    num1 = int(data[0][i])
    num2 = int(data[1][i])
    num3 = int(data[2][i])
    num4 = int(data[3][i])
    operator = data[4][i]

    if operator == '*':
        result = num1 * num2 * num3 * num4
    elif operator == '+':
        result = num1 + num2 + num3 + num4
    else:
        raise ValueError(f"Unknown operator: {operator}")

    calculations += result
    
end_time = perf_counter_ns()
print(f"Total calculations result: {calculations}")
print(f"Execution time: {(end_time - start_time) / 1_000_000} ms")
```

## Part 2

```python
start_time = perf_counter_ns()

with open("input.txt") as f:
    lines = f.read().splitlines()

calculations = 0
operator_line = lines[-1]
number_lines = lines[:-1]

i = 0
last_col_end = -1
while i < len(operator_line):
    operator = operator_line[i]
    
    col_start = max(last_col_end + 1, 0)
    
    col_end = i
    # operator is always at the start of the column
    while col_end < len(operator_line) - 1 and operator_line[col_end + 1] == ' ':
        col_end += 1
    
    numbers = []
    for char_pos in range(col_start, col_end + 1):
        digits = ''.join(line[char_pos] if char_pos < len(line) and line[char_pos].isdigit() else '' for line in number_lines)
        if digits:
            numbers.append(int(digits))
    
    if operator == '*':
        result = 1
        for num in numbers:
            result *= num
    else:
        result = sum(numbers)

    
    calculations += result
    last_col_end = col_end
    i = col_end + 1

end_time = perf_counter_ns()
print(f"Total calculations result: {calculations}")
print(f"Execution time: {(end_time - start_time) / 1_000_000} ms")
```
