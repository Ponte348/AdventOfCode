```python
from time import perf_counter_ns

with open("input.txt") as f:
    data = [list(line) for line in f.read().splitlines()]
```

## Part 1

```python
start_time = perf_counter_ns()

splits=0

for i in range(len(data[0])):
    if data[0][i] == 'S':
        data[1][i] = '|'
        break

for row in range(1, len(data)):
    for col in range(len(data[0])):
        if data[row][col] == '^' and data[row-1][col] == '|':
            splits += 1
            data[row][col-1] = '|'
            data[row][col+1] = '|'
        elif data[row][col] == '.' and data[row-1][col] == '|':
            data[row][col] = '|'

end_time = perf_counter_ns()
print("Beams was split:", splits, "times")
print(f"Execution time: {(end_time - start_time) / 1_000_000} ms")
```
