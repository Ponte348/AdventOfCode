```python
from time import perf_counter_ns

with open("input.txt") as f:
    data = [list(line) for line in f.read().splitlines()]
```

## Part 1

```python
start_time = perf_counter_ns()

splits=0

for i in range(len(data[0])):
    if data[0][i] == 'S':
        data[1][i] = '|'
        break

for row in range(1, len(data)):
    for col in range(len(data[0])):
        if data[row][col] == '^' and data[row-1][col] == '|':
            splits += 1
            data[row][col-1] = '|'
            data[row][col+1] = '|'
        elif data[row][col] == '.' and data[row-1][col] == '|':
            data[row][col] = '|'
            
end_time = perf_counter_ns()
print("Beams was split:", splits, "times")
print(f"Execution time: {(end_time - start_time) / 1_000_000} ms")
```

## Part 2

```python
from time import perf_counter_ns

with open("input.txt") as f:
    data = [list(line) for line in f.read().splitlines()]
```

```python
start_time = perf_counter_ns()

for i in range(len(data[0])):
    if data[0][i] == 'S':
        data[1][i] = '|'
        break

for row in range(1, len(data)):
    for col in range(len(data[0])):
        if data[row][col] == '^' and data[row-1][col] == '|':
            data[row][col-1] = '|'
            data[row][col+1] = '|'
        elif data[row][col] == '.' and data[row-1][col] == '|':
            data[row][col] = '|'

for i in range(len(data[0])):
    if data[0][i] == 'S':
        data[1][i] = 1
        break

for row in range(len(data)):
    for col in range(len(data[0])):
        
        if data[row][col] == '|':
            data[row][col] = data[row-1][col] if isinstance(data[row-1][col], int) else 0
            left, right = 0, 0

            if col-1 >= 0 and data[row][col-1] == '^' and isinstance(data[row-1][col-1], int):
                left = int(data[row-1][col-1])
            if col+1 < len(data[0]) and data[row][col+1] == '^' and isinstance(data[row-1][col+1], int):
                right = int(data[row-1][col+1])
            data[row][col] += left + right
            
            if col-1 >= 0 and col+1 < len(data[0]) and data[row][col-1] != '^' and data[row][col+1] != '^':
                data[row][col] = data[row-1][col]

# sum all digits in the last row
timelines = sum(int(x) for x in data[-1] if isinstance(x, int))
end_time = perf_counter_ns()
print("Number of timelines:", timelines)
print(f"Execution time: {(end_time - start_time) / 1_000_000} ms")
```
