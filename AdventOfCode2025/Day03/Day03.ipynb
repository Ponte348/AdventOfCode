```python
from time import perf_counter_ns
from itertools import combinations

with open("input.txt") as f:
    data = f.read().splitlines()
data = [[int(x) for x in line] for line in data]
```

## Part 1

```python
start_time = perf_counter_ns()

total_output_voltage = 0

for array in data:
    max_concatenation = 0
    for i in range(len(array) - 1):
        for j in range(i + 1, len(array)):
            concatenated = int(str(array[i]) + str(array[j]))
            if concatenated > max_concatenation:
                max_concatenation = concatenated
    total_output_voltage += max_concatenation

end_time = perf_counter_ns()
print("Total output voltage:", total_output_voltage)
print(f"Execution time: {(end_time - start_time) / 1_000_000} ms")
```

## Part 2

```python
start_time = perf_counter_ns()

total_output_voltage = 0
for array in data:
    result = []
    start_pos = 0
    for digit_num in range(12):
        remaining = 12 - digit_num
        max_digit = -1
        max_pos = -1
        for i in range(start_pos, len(array) - remaining + 1):
            if array[i] > max_digit:
                max_digit = array[i]
                max_pos = i
        result.append(max_digit)
        start_pos = max_pos + 1
    max_concatenation = int(''.join(str(d) for d in result))
    total_output_voltage += max_concatenation

    
end_time = perf_counter_ns()
print("Total output voltage:", total_output_voltage)
print(f"Execution time: {(end_time - start_time) / 1_000_000} ms")
```
