# Day 1

```python
from time import perf_counter_ns

# read file "input.txt", each line has the format XNN where X is either "L" or "R" and NN is a two digit number
with open("input.txt") as f:
    lines = f.read().strip().split("\n")
directions = [(line[0], int(line[1:])) for line in lines]
#print(directions)

# read "example.txt"
with open("example.txt") as f:
    lines = f.read().strip().split("\n")
example_directions = [(line[0], int(line[1:])) for line in lines]
print(example_directions)
```

## Part 1

```python
start_time = perf_counter_ns()

counter = 50
zeroes = 0

for direction, value in directions:
    if direction == "L":
        counter -= value
    else:
        counter += value
    counter %= 100
    if counter == 0:
        zeroes += 1

end_time = perf_counter_ns()
print(f"Password: {zeroes}")
print(f"Execution time: {(end_time - start_time) / 1_000_000} ms")
```

## Part 2

```python
start_time = perf_counter_ns()

counter = 50
zeroes = 0

for direction, value in directions:
    for i in range(1, value+1):
        if direction == "L":
            counter -= 1
        else:
            counter += 1
        counter %= 100
        if counter == 0:
            zeroes += 1
        

end_time = perf_counter_ns()
print(f"Password: {zeroes}")
print(f"Execution time: {(end_time - start_time) / 1_000_000} ms")
```
