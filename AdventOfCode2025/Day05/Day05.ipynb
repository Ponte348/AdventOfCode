# Day 05

```python
from time import perf_counter_ns

with open("input.txt") as f:
    data = f.read().strip().split("\n\n")
ranges_data, ids_data = data
ranges = [line.split("-") for line in ranges_data.split("\n")]
ids = [int(line) for line in ids_data.split("\n")]
```

## Part 1

```python
start_time = perf_counter_ns()

count = 0

for id_ in ids:
    valid = False
    for r in ranges:
        low, high = map(int, r)
        if low <= id_ <= high:
            valid = True
            break
    if valid:
        count += 1
        
end_time = perf_counter_ns()
print("Fresh ingredient IDs:", count)
print(f"Execution time: {(end_time - start_time) / 1_000_000} ms")
```

## Part 2

```python
start_time = perf_counter_ns()

merged_ranges = []
for r in sorted(ranges, key=lambda x: int(x[0])):
    low, high = map(int, r)
    if not merged_ranges or merged_ranges[-1][1] < low - 1:
        merged_ranges.append([low, high])
    else:
        merged_ranges[-1][1] = max(merged_ranges[-1][1], high)
        
count = 0
for r in merged_ranges:
    low, high = r
    count += (high - low + 1)


end_time = perf_counter_ns()
print("Total fresh ingredient IDs:", count)
print(f"Execution time: {(end_time - start_time) / 1_000_000} ms")
```
